```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'merged_state_coverage_data3 (2).xlsx'
df = pd.read_excel(file_path)

# Select relevant columns
columns = [
    'State',
    '2010 Youth Employer Based Coverage Percentage',
    'Youth Any Coverage 2010',
    'Youth 2015 Employer Based Coverage Percentage',
    'Youth Any Coverage 2015',
    'Youth 2021 Employer Based Coverage Percentage',
    'Youth Any Coverage 2021'
]
df = df[columns].copy()

# Calculate differences and sort states based on 2010 coverage
df['Coverage Change (2010-2021)'] = df['Youth 2021 Employer Based Coverage Percentage'] - df['2010 Youth Employer Based Coverage Percentage']
high_coverage_states = df[df['2010 Youth Employer Based Coverage Percentage'] >= df['2010 Youth Employer Based Coverage Percentage'].median()]
low_coverage_states = df[df['2010 Youth Employer Based Coverage Percentage'] < df['2010 Youth Employer Based Coverage Percentage'].median()]

# Plot trends for high vs. low initial coverage
plt.figure(figsize=(12, 8))

# High coverage in 2010
plt.subplot(2, 1, 1)
sns.barplot(data=high_coverage_states, x='State', y='Coverage Change (2010-2021)', palette='Blues')
plt.title('Coverage Change for States with High Youth Coverage in 2010')
plt.xticks(rotation=90)

# Low coverage in 2010
plt.subplot(2, 1, 2)
sns.barplot(data=low_coverage_states, x='State', y='Coverage Change (2010-2021)', palette='Greens')
plt.title('Coverage Change for States with Low Youth Coverage in 2010')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
