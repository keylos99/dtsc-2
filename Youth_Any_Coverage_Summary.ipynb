```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = 'merged_state_coverage_data3 (2).xlsx'
data = pd.read_excel(file_path)

# Select Any coverage columns
columns = [
    'State',
    'Youth Any Coverage 2010',
    'Youth Any Coverage 2015',
    'Youth Any Coverage 2021'
]
df = data[columns].copy()

# Calculate changes over time
df['Coverage Change (2010-2015)'] = df['Youth Any Coverage 2015'] - df['Youth Any Coverage 2010']
df['Coverage Change (2015-2021)'] = df['Youth Any Coverage 2021'] - df['Youth Any Coverage 2015']
df['Total Coverage Change (2010-2021)'] = df['Youth Any Coverage 2021'] - df['Youth Any Coverage 2010']

# Plot changes over time
plt.figure(figsize=(12, 10))

# Change from 2010 to 2015 Plot
plt.subplot(3, 1, 1)
sns.barplot(data=df, x='State', y='Coverage Change (2010-2015)', palette='Purples')
plt.title('Youth Any Coverage Change (2010-2015)')
plt.xticks(rotation=90)

# Change from 2015 to 2021 Plot
plt.subplot(3, 1, 2)
sns.barplot(data=df, x='State', y='Coverage Change (2015-2021)', palette='Oranges')
plt.title('Youth Any Coverage Change (2015-2021)')
plt.xticks(rotation=90)

# Total change from 2010 to 2021 Plot
plt.subplot(3, 1, 3)
sns.barplot(data=df, x='State', y='Total Coverage Change (2010-2021)', palette='Greens')
plt.title('Youth Any Coverage Change (2010-2021)')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
